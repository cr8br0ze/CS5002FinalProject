```python
def evaluate(text):
    text = text.translate(str.maketrans('', '', string.punctuation)).split()
    pSpamText = 0.13
    for i in text:
        if i.lower() in pTextSpam.keys():
            pSpamText *= pTextSpam[i.lower()]
        else:
            pSpamText *= (1)/(len(trainingDic))
    return pSpamText
```
